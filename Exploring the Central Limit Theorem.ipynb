```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import ttest_ind
```

```python
# creating 2 populatians
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

# plotting a histogram for both populations
plt.hist(pop1, color = 'c', alpha = 0.5, label = 'Population 1 (p = 0.2)')
plt.hist(pop2, color = 'g', alpha = 0.5, label = 'Population 2')
plt.legend(loc = 'best')
plt.show()
```

```python
# bootstrap 1000 samples from each population
sample1 = np.random.choice(pop1, size = 1000, replace = True)
sample2 = np.random.choice(pop2, size = 1000, replace = True)

# plotting a histogram for both samples
plt.hist(sample1, color = 'c', alpha = 0.5, label = 'Sample 1')
plt.hist(sample2, color = 'g', alpha = 0.5, label = 'Sample 2')
plt.legend(loc = 'best')
plt.show()

# printing mean and standard deviation for both samples
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
print('Difference of means: {}'.format(sample2.mean()-sample1.mean()))
```

```python
# changing bootstrap sizes to 20 samples
sample1 = np.random.choice(pop1, size = 20, replace = True)
sample2 = np.random.choice(pop2, size = 20, replace = True)

# plotting a histogram for both samples
plt.hist(sample1, color = 'orange', alpha = 0.5, label = 'New Sample 1')
plt.hist(sample2, color = 'r', alpha = 0.5, label = 'New Sample 2')
plt.legend(loc = 'best')
plt.show()

# printing mean and standard deviation for both samples
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

print('As sample sizes decreased, means changed more, but standard deviations have little changes')
print('With the smaller sample sizes, sample means have less tendency to follow a normal probability distribution')
print('Sample means do not cluster around true population means')
```

```python
# changing the probability value for pop1 to 0.3
pop1 = np.random.binomial(10, 0.3, 10000)

# plotting a histogram for both populations
plt.hist(pop1, color = 'y', alpha = 0.5, label = 'Population 1 (p = 0.3)')
plt.hist(pop2, color = 'b', alpha = 0.5, label = 'Population 2')
plt.legend(loc = 'best')
plt.show()

# bootstrap 100 samples from each population
sample1 = np.random.choice(pop1, size = 100, replace = True)
sample2 = np.random.choice(pop2, size = 100, replace = True)


# calculating the t-value and p-value
print(ttest_ind(sample1, sample2, equal_var = False))
```

```python
# changing the probability value for pop1 to 0.4
pop1 = np.random.binomial(10, 0.4, 10000)

# plotting a histogram for both populations
plt.hist(pop1, color = 'orange', alpha = 0.5, label = 'Population 1 (p = 0.4)')
plt.hist(pop2, color = 'purple', alpha = 0.5, label = 'Population 2')
plt.legend(loc = 'best')
plt.show()

# bootstrap 100 samples from each population
sample1 = np.random.choice(pop1, size = 100, replace = True)
sample2 = np.random.choice(pop2, size = 100, replace = True)

# difference of means
print('Difference of means: {}'.format(sample2.mean() - sample1.mean()))

# calculating the t-value and p-value
print(ttest_ind(sample1, sample2, equal_var = False))


print('Change of probability from 0.3 to 0.4 decreased the t-value as well as p-value')
print('Low variability can lead to groups with means that look very similar')
print('The last p-value is lower showing that the difference in means is due to a real difference in the population and not due to variability in the samples')
```

```python
# changing distribution model
pop1 = np.random.gamma(10, 0.2, 10000)
pop2 = np.random.gamma(10, 0.5, 10000)

# plotting a histogram for both Gamma distributed populations
plt.hist(pop1, color = 'black', alpha = 0.5, label = "Population 1 (Gamma)")
plt.hist(pop2, color = 'red', alpha = 0.5, label = "Population 2 (Gamma)")
plt.legend(loc='best')
plt.show()

# bootstrap 100 samples from each population
sample1 = np.random.choice(pop1, size = 100, replace = True)
sample2 = np.random.choice(pop2, size = 100, replace = True)

print('Sample 1 mean: {}'. format(sample1.mean()))
print('Sample 2 mean: {}'. format(sample2.mean()))
print('Difference of means: {}'.format(sample2.mean()-sample1.mean()))

print(ttest_ind(sample1, sample2, equal_var = False))
```
